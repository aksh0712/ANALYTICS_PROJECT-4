```python
def translator(text):
    '''
    Custom function to translate a foreign language to English
    '''
    from googletrans import Translator, constants
    translator = Translator()
    src_lang = translator.detect(text).lang    
    if src_lang != 'en':
        translation = translator.translate(text)
        op = translation.text 
    else:
        op=text
    return op
```

## 2. Removing Punctuations

```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["NewsBlurb"] = df["NewsBlurb"].apply(lambda text: remove_punctuation(text))
df.head()
```

## 3. Removing StopWords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
print(", ".join(stopwords.words('english')))
```

```python
STOPWORDS = set(stopwords.words('english'))
STOPWORDS.add("·") # · is different from .

some_unwanted_words = ['say','small','volkswagens','renault','rover','rollsroyce','jaguar','mahindra','suzuki','join','tata','maruti',
                       'mitsubishi','bridgestone','michelin','continental','nissan','subaru','nois','porche','audi','september'
                      ]
STOPWORDS = set(list(STOPWORDS) + some_unwanted_words)
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["NewsBlurb"] = df["NewsBlurb"].apply(lambda text: remove_stopwords(text))
df.head()
```

## 4. Dealing with Frequent but Unwanted Words

```python
from collections import Counter
cnt = Counter()
for text in df["NewsBlurb"].values:
    for word in text.split():
        cnt[word] += 1
        
all_words_freq = pd.Series(list(cnt.values()), index = cnt.keys()).sort_values(ascending=False)
all_words_freq
```

```python
all_words_freq[0:120].plot(kind="bar",figsize=(15,8),color="#EF2D56")
plt.show()
```

```python
from wordcloud import WordCloud
text = all_words_freq.values
wordcloud = WordCloud(background_color = "white", max_words = 100,width=1440, height=700).generate_from_frequencies(all_words_freq)
plt.figure(figsize = (15, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0) 

plt.show()
```

```python
# remove the 100 most common of them
FREQWORDS = set([w for (w, wc) in cnt.most_common(100)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df["NewsBlurb"] = df["NewsBlurb"].apply(lambda text: remove_freqwords(text))
df.head()
```

## 5. Deleting Rarely Occuring Words

```python
all_words_freq
```

```python
n_rare_words = 20
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

df["NewsBlurb"] = df["NewsBlurb"].apply(lambda text: remove_rarewords(text))
df.head()
```

## 6. Removing Inflected Words by Lemmatization

Can use **Stemming**

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

df["NewsBlurb"] = df["NewsBlurb"].apply(lambda text: lemmatize_words(text))
df.head()
```

## 7. Removing URLs, HTML Tags

```python
def remove_urls(text):
    import re
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)
```

```python
nb = df["NewsBlurb"].apply(lambda text: remove_urls(text))
nb = df["NewsBlurb"].apply(lambda text: remove_html(text))
df["NewsBlurb"] = nb
df.head()
```

```python
from collections import Counter
# from scipy.misc import imread
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
from matplotlib import pyplot as plt
%matplotlib inline
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
```

```python
def print_top_words(model, feature_names, n_top_words):
    '''
    Custom helper function to print top words
    '''
    for index, topic in enumerate(model.components_):
        message = "\nTopic #{}:".format(index)
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 :-1]])
        print(message)
        print("="*70)
```

```python
class LemmaCountVectorizer(CountVectorizer):
    '''
    Custom class for Creating CountVectorizer using the Lemmas
    '''
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
```

```python
# Storing the entire training text in a list
text = list(df.NewsBlurb.values)
# Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, 
                                     min_df=2,
                                     stop_words='english',
                                     decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

```python
feature_names = tf_vectorizer.get_feature_names()
count_vec = np.asarray(tf.sum(axis=0)).ravel()
zipped = list(zip(feature_names, count_vec))
x, y = (list(x) for x in zip(*sorted(zipped, key=lambda x: x[1], reverse=True)))

# Now I want to extract out on the top 15 and bottom 15 words
Y = np.concatenate([y[0:15], y[-16:-1]])
X = np.concatenate([x[0:15], x[-16:-1]])
```

#### Let's see what happens when we use a spaCy based bigram tokenizer for topic modelling

```python
def spacy_bigram_tokenizer(phrase):
    doc = parser(phrase) # create spacy object
    token_not_noun = []
    notnoun_noun_list = []
    noun = ""

    for item in doc:
        if item.pos_ != "NOUN": # separate nouns and not nouns
            token_not_noun.append(item.text)
        if item.pos_ == "NOUN":
            noun = item.text
        
        for notnoun in token_not_noun:
            notnoun_noun_list.append(notnoun + " " + noun)

    return " ".join([i for i in notnoun_noun_list])
```

```python
def sentiment_scores(sentence): 
    
    '''
    Custom function to calculate the sentiment scores.
    '''
  
    # Create a SentimentIntensityAnalyzer object. 
    sid_obj = SentimentIntensityAnalyzer() 
  
    # polarity_scores method of SentimentIntensityAnalyzer 
    # oject gives a sentiment dictionary. 
    # which contains pos, neg, neu, and compound scores. 
    sentiment_dict = sid_obj.polarity_scores(sentence) 
    
    # decide sentiment as positive, negative and neutral 
    if sentiment_dict['compound'] >= 0.05 : 
        senti = "POSITIVE" 
  
    elif sentiment_dict['compound'] <= - 0.05 : 
        senti = "RISK" 
  
    else : 
        senti = "NEUTRAL"
    return senti
```

```python
def sentiment_scores_calc(sentence): 

    # Create a SentimentIntensityAnalyzer object. 
    sid_obj = SentimentIntensityAnalyzer() 

    # polarity_scores method of SentimentIntensityAnalyzer 
    # oject gives a sentiment dictionary. 
    # which contains pos, neg, neu, and compound scores. 
    sentiment_dict = sid_obj.polarity_scores(sentence) 


    senti = [str(round(sentiment_dict['compound']*100,2))]
    score = ", ".join(senti)
    return(score)
    
```

```python
sentiment = news_blurb.apply(lambda x:sentiment_scores(x))
```

```python
auto_suppliers["SENTIMENT"] = sentiment
```

```python
sent_score = news_blurb.apply(lambda x:sentiment_scores_calc(x))
```

```python
auto_suppliers["NEWS_SCORE"] = sent_score
```
